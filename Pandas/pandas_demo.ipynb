```python
import pandas as pd
```

```python
# Series in pandas
s = pd.Series([23, 24, 25, 26])
print(s)
print(type(s))
```

```python
# Indexing
print(s[0])    # 23
print(s[2])    # 25

print(s.index)     # all label
```

```python
# Custom Indexing
s2 = pd.Series([23, 24, 25, 26], index = ["Adam", "Eve", "Charlie", "Bob"])
print(s2["Eve"])    # 24
print(s2["Bob"])    # 26
```

```python
# Vectorized Operations
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])

print(s1 + s2)
```

```python
# Mutable Values but immutable size
s = pd.Series([1, 2, 3, 4, 5])
s[0] = 100

print(s)
```

```python
changed_s = s.drop(1)
print(changed_s)
print(s)
```
