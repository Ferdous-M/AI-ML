```python
# Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set Theme
sns.set_theme() # sets default theme - whitegrid, darkdark, white, dark, ticks
sns.get_dataset_names()
```

```python
# Load Data
tips = sns.load_dataset("tips")
print(type(tips))
tips.head()
tips.tail()
```

```python
# Plot
sns.set_theme(style="white")
sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    hue="smoker", style="smoker", size="size",
)
```
