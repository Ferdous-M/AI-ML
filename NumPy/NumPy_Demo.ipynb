```python
import numpy as np
import time
```

```python
#execution time differnece
size = 1_000
py_list = list(range(size))
#print(py_list)

start = time.time()
sq_list = [x**2 for x in py_list]
end =  time.time()
print(f"python list time = {end - start} seconds")

np_arr = np.array(py_list)
#print(np_arr)

start = time.time()
sq_arr = np_arr ** 2
end =  time.time()
print(f"numpy array time = {end - start} seconds")
```

```python
#create with built in functions
arr1 = np.zeros((2,3),dtype = "int64")
print(arr1, arr1.shape)

arr1 = np.ones((5,))
print(arr1, arr1.shape)

arr3 = np.full((3,3),100)
print(arr3, arr3.shape)

arr4 = np.eye(3)
print(arr4, arr4.shape)

arr5 = np.arange(0,11,2)
print(arr5, arr5.shape)

arr5 = np.linspace(1,100,4)
print(arr5, arr5.shape)
```

```python
#properties
arr = np.array([[1,2,3],[6,4,5]])
print(arr.shape)

reshaped = arr.reshape((3,2))

print(reshaped, reshaped.size)
```
